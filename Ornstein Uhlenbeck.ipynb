```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T=1
N=500
M=100
h=T/N
theta=1.0 #tester pour theta = 10 
mu=10.0
sigma=3
X0=0

# XT=N(u(1-e^{-theta}),sigma^2)

t=np.linspace(0,T,N+1, endpoint=True)
Wc = np.zeros((N+1,M))
Xc = np.zeros((N+1,M))
W0=0
dWc=np.sqrt(h)*np.random.randn(N,M)
Wc[0,:]=0
Wc[1:,:]=np.cumsum(dWc,0)

for i in range(N):
    
    Xc[i+1]=Xc[i]+theta*(mu-Xc[i])*h+sigma*(Wc[i+1]-Wc[i])
    
plt.figure(figsize=(16,9))
plt.plot(t,Xc)
plt.title('Ornstein Uhlenbeck')
plt.grid()
plt.show()
```

```python

plt.hist(Xc[N], bins=40, facecolor='red',edgecolor='black', alpha=0.7)
```
